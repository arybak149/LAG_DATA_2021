```python
# data upload and exploration
df = pd.read_csv('https://raw.githubusercontent.com/CunyLaguardiaDataAnalytics/datasets/master/2014-15_To_2016-17_School-_Level_NYC_Regents_Report_For_All_Variables.csv')
df.head(2)
```

```python
#School level selection = HS and dropping NA columns 
All_HS_df = df[ df['School Level'] == 'High school']
All_HS_df.drop(['Number Scoring CR', 'Percent Scoring CR'], axis = 1).head(3)
```

```python
#removing rows that have 's'
All_HS_df=All_HS_df[All_HS_df['Mean Score'] !='s']
display(All_HS_df)
```

```python

All_HS_df[['Mean Score']] = All_HS_df[['Mean Score']].apply(pd.to_numeric)
display(All_HS_df)
```

```python
# Regents Exam -> Subject
# Total Tested -> Students Tested
# School Name -> Name
# School Level -> Level
# School DBN -> DBN
All_HS_df.rename(columns={'Regents Exam': 'Subject', 'Total Tested': 'Students Tested', 'School Name' : 'Name', 'School Level' : 'Level', 'School DBN' : 'DBN'}, inplace=True)
All_HS_df.head(4)
```
